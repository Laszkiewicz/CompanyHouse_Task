```python
import pandas as pd
pd.options.mode.chained_assignment = None  # default='warn'
```

```python
data = {'Provincce': ['dolnoslaskie', 'kujawsko-pomorskie', 'lubelskie', 'lubuskie', 'lodzkie', 'malopolskie', 'mazowieckie'],
        'Register no.': ['2', '4', '6', '8', '10', '12', '14'],
        'Provincial city':[ 'wroclaw', 'bydgoszcz', 'lublin', 'gorzow wielkopolski / zielona Gora', 'Lodz', 'Krakow', 'Warszawa'],
        'Area [km^2]':[19946.70, 17971.34, 25122.46, 13987.93, 18218.95, 15182.79, 35558.47],
        'Population': [2904198,2086210, 139726, 1018084, 2493603, 3372618, 5349114]
       }
df = pd.DataFrame(data)
```

<h1> Showing DF from previously set up data and dataframe </h1>

```python
df
```

<h2> Listing data to area less than 20000  </h2>

```python
df.loc[df['Area [km^2]'] <20000]
```

<h2>Setting Population which exceedes 2000000 </h2>

```python
new_df = df.loc[df['Population'] > 2000000]
new_df
```

<h2> Adding new row as pos 0 </h2>

```python
new_df.loc[0] = ['wielkopolskie', '30', 'Poznan', 29826.50, 3475232]
```

```python
new_df
```

<h2> Sorting Values by Population descending</h2>

```python
new_df = new_df.sort_values('Population', ascending=False)
```

<h2> Moving column Provincial city to the end </h2>

```python
new_df = new_df[['Provincce', 'Register no.', 'Area [km^2]', 'Population', 'Provincial city']]
new_df
```

<h2> Setting Provincial City as capitalised </h2>

```python
new_df['Provincial city'] = new_df['Provincial city'].str.capitalize()
new_df
```

```python
new_df['Density'] = new_df['Population'] / new_df['Area [km^2]']

new_df
```
